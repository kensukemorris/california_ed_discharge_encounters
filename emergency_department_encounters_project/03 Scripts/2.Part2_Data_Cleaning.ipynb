```python
# Check for mixed variable types within columns

mixed_list = []

for col in df.columns:
    mixed_type = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
    if len(df[mixed_type]) > 0:
        print(col)
        mixed_list.append(col)
```

```python
# change column type for all mixed type variables

for col in mixed_list:
    df[col] = df[col].astype('str')
```

```python
#Since there are mixed type variables, the column cannot be converted to a string. The code below seperated the string that
# that can be converted to floats, then the floats are converted to a categorical string.

prev_beds = ['Beds 1-49','Beds 50-99','Beds 100-149','Beds 150-199','Beds 200-299',
             'Beds 300-499','Beds 500+']
bed_cat = ['1-49','50-99','100-149','150-199','200-299','300-499','500+']

s = 'Beds'
t = '-'
u = '+'

for val in df['LICENSED_BED_SIZE']:
        if s in str(val):
            if val == prev_beds[0] :
                df.loc[df['LICENSED_BED_SIZE'] == val,'LICENSED_BED_SIZE'] = bed_cat[0]
                print(val)
            elif val == prev_beds[1]:
                df.loc[df['LICENSED_BED_SIZE'] == val,'LICENSED_BED_SIZE'] = bed_cat[1]
                print(val)
            elif val == prev_beds[2]:
                df.loc[df['LICENSED_BED_SIZE'] == val,'LICENSED_BED_SIZE'] = bed_cat[2]
                print(val)
            elif val == prev_beds[3]:
                df.loc[df['LICENSED_BED_SIZE'] == val,'LICENSED_BED_SIZE'] = bed_cat[3]
                print(val)
            elif val == prev_beds[4]:
                df.loc[df['LICENSED_BED_SIZE'] == val,'LICENSED_BED_SIZE'] = bed_cat[4]
                print(val)
            elif val == prev_beds[5]:
                df.loc[df['LICENSED_BED_SIZE'] == val,'LICENSED_BED_SIZE'] = bed_cat[5]
                print(val)
            elif val == prev_beds[6]:
                df.loc[df['LICENSED_BED_SIZE'] == val,'LICENSED_BED_SIZE'] = bed_cat[6]
                print(val)
        elif t in str(val):
            pass
        elif u in str(val):
            pass
        else:
            if (float(val) in range(1,50)):
                df.loc[df['LICENSED_BED_SIZE'] == val,'LICENSED_BED_SIZE'] = bed_cat[0]
                print(val)
            elif float(val) in range(50,100):
                df.loc[df['LICENSED_BED_SIZE'] == val,'LICENSED_BED_SIZE'] = bed_cat[1]
                print(val)
            elif float(val) in range(100,150):
                df.loc[df['LICENSED_BED_SIZE'] == val,'LICENSED_BED_SIZE'] = bed_cat[2]
                print(val)
            elif float(val) in range(150,200):
                df.loc[df['LICENSED_BED_SIZE'] == val,'LICENSED_BED_SIZE'] = bed_cat[3]
                print(val)
            elif float(val) in range(200,300):
                df.loc[df['LICENSED_BED_SIZE'] == val,'LICENSED_BED_SIZE'] = bed_cat[4]
                print(val)
            elif float(val) in range(300,500):
                df.loc[df['LICENSED_BED_SIZE'] == val,'LICENSED_BED_SIZE'] = bed_cat[5]
                print(val)
            elif float(val) > 500:
                df.loc[df['LICENSED_BED_SIZE'] == val,'LICENSED_BED_SIZE'] = bed_cat[6]
                print(val)
            
```

# 3. Dealing with Missing Values

```python
#load previously save dataset. USe excel to change'nan' to blank

df_updated = pd.read_csv(os.path.join(data_path,'Prepared', '1.data_mix_check.csv'))
df_updated.tail()
```

```python
#For license category description, imputed missing value for one of the facilities in excel by searching for similar id. 
#Will delete the remaining two values in jupyter
```

```python
dict(df_updated.isnull().sum())
```

```python
# Dropping NA in categorical columns that are central to my curiostity

df_updated.dropna(axis = 0, how = 'any', subset = ['er_service_level_desc', 'license_category_desc',
                                                   'licensed_bed_size','sex_male'], inplace = True)
```

```python
#remove missing values in license_category_desc column = 0
df_updated = df_updated[df_updated['licensed_bed_size'] != '0']

df_updated['licensed_bed_size'].value_counts(dropna = False)
```

```python
#replace na values for teaching hospitals based on assumption, along with other numeric columns where the likely hood
# having a value of zero is high

df_updated[['teaching_hospital_desc','sex_unk','age_unknown']] = df_updated[['teaching_hospital_desc',
                                                                            'sex_unk','age_unknown']].fillna(0)
# df_updated['teaching_hospital_desc'].value_counts(dropna = False)
```
